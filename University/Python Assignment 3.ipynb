## Creating a statistical dataframe.

```python
d={'no_of_present':0,'less than 50%':0,'between 50% and 60%':0,'between 60% and 70%':0,'between 70% and 80%':0,'greater than 80%':0}
stat_frame=pd.DataFrame(d,index=(0,1))
stat_frame
```

Inference- Creating a new data frame where we will save the counts of the marks.

```python
temp={1:bw1,2:bw2,3:bw3,4:bw4,5:bw5,6:bw6,7:bw7,8:bw8,9:bw9,10:bw10,11:bw11,12:bw12,13:bw13,14:bw14,15:bw15,16:bw16,17:bw17,18:bw18,19:bw19}
result={}

for i in temp:
    result[i]=pd.merge(final_quiz,temp[i],on='studentName',how='inner')
```

```python
result[15]
```

Inference- Creating a for loop where we give the batch number and it gives out the entire data set merged with the quiz marks.

```python
stats={}
d={'no_of_present':0,'less than 50%':0,'between 50% and 60%':0,'between 60% and 70%':0,'between 70% and 80%':0,'greater than 80%':0}
stat_frame=pd.DataFrame(d,index=(0,1))
stat_frame

for i in result :
    stat_frame=pd.DataFrame(d,index=(0,1))
    stat_frame.loc[0,'no_of_present']=result[i].Quiz_No_x.count()
    stat_frame.loc[1,'no_of_present']=result[i].Quiz_No_y.count()
    stat_frame.loc[0,'between 50% and 60%']=result[i]['Grade/10.00_x'].between(5.0,5.9).sum()
    stat_frame.loc[1,'between 50% and 60%']=result[i]['Grade/10.00_y'].between(5.0,5.9).sum()
    stat_frame.loc[0,'between 60% and 70%']=result[i]['Grade/10.00_x'].between(6.0,6.9).sum()
    stat_frame.loc[1,'between 60% and 70%']=result[i]['Grade/10.00_y'].between(6.0,6.9).sum()
    stat_frame.loc[0,'between 70% and 80%']=result[i]['Grade/10.00_x'].between(7.0,7.9).sum()
    stat_frame.loc[1,'between 70% and 80%']=result[i]['Grade/10.00_y'].between(7.0,7.9).sum()
    x=result[i][result[i]['Grade/10.00_x']>=8.0].count(axis=0)
    x=x[1]
    y=result[i][result[i]['Grade/10.00_y']>=8.0].count(axis=0)
    y=y[1]
    stat_frame.loc[0,'greater than 80%']=x
    stat_frame.loc[1,'greater than 80%']=y
    p=result[i][result[i]['Grade/10.00_x']<5].count(axis=0)
    p=p[1]
    q=result[i][result[i]['Grade/10.00_y']<5].count(axis=0)
    q=q[1]
    stat_frame.loc[0,'less than 50%']=p
    stat_frame.loc[1,'less than 50%']=q
    stat_frame['Batch']=i
    stat_frame['Quiz Number']=(1,2)
    
    stats[i]=stat_frame
    

    
```

Inference- The statistic data frame which was created is now filled using the for loop and we get the above output.
